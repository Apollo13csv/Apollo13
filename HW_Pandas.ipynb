```python
data = pd.read_csv(r'C:\Users\Apollo13\Desktop\ml-latest\ratings.csv')
data.head(10)
```

```python
rating = data.filter(items = ['rating', 'movieId'])
```

```python
max_rating = rating['rating'].max()
rating_joined = rating[(rating ['rating'] == max_rating)]
rating_f = rating_joined.groupby(['movieId'])['movieId'].count()
```
